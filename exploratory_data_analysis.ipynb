# Import Python Packages

```python
import pandas as pd
import numpy as np
import librosa as lbr
import matplotlib.pyplot as plt
```

# Inspect the Audio

```python
# Load audio file
audio_data, sampling_rate = lbr.load('data/en001a.wav', sr=None) #sr represents the sampling rate
```

```python
# Get the duration of the audio in seconds 
duration = lbr.get_duration(y=audio_data, sr=sampling_rate)
print(str(duration) + " seconds")
```

```python
# Visualize the audio waveform
plt.figure(figsize=(12, 4))

lbr.display.waveshow(audio_data, sr=sampling_rate)

plt.title('Waveform of Audio Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.show()
```

# Feature Extraction

```python
# Mel-spectrogram represents the spectral content of the audio signal
plt.figure(figsize=(10, 4))

mel_spectrogram = lbr.feature.melspectrogram(y=audio_data, sr=sampling_rate)
lbr.display.specshow(lbr.power_to_db(mel_spectrogram, ref=np.max), sr=sampling_rate, hop_length=512)

plt.colorbar(format='%+2.0f dB')
plt.title("Mel-spectrogram")
plt.xlabel('Time')
plt.ylabel('Hz')
plt.show()
```

```python
# Mel Frequency Cepstral Coefficients for speech/music analysis
plt.figure(figsize=(10, 4))

mfcc= lbr.feature.mfcc(y=audio_data, sr=sampling_rate, n_mfcc=13)
lbr.display.specshow(mfcc, y_axis='mel', x_axis='time')

plt.colorbar(format='%+2.0f dB')

plt.title('MFCCs')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficient')
plt.show()
```

```python
# Chroma features for capturing the harmonic progression of an audio signal
plt.figure(figsize=(10, 4))

chroma = lbr.feature.chroma_cqt(y=audio_data, sr=sampling_rate)
lbr.display.specshow(chroma, y_axis='chroma', x_axis='time')

plt.colorbar()

plt.title('Chromagram')
plt.xlabel('Time')
plt.ylabel('Pitch Class')
plt.show()
```

```python
# Spectral contrast features for capturing the regions of high spectral activity

plt.figure(figsize=(10, 4))

contrast = lbr.feature.spectral_contrast(y=audio_data, sr=sampling_rate)
lbr.display.specshow(contrast, x_axis='time')

plt.colorbar()

plt.title('Spectral Contrast')
plt.xlabel('Time')
plt.ylabel('Frequency Band')
plt.show()
```

```python
# Tonnetz features for mapping the chroma features into a six-dimensional space

plt.figure(figsize=(10, 4))

tonnetz = lbr.feature.tonnetz(y=audio_data, sr=sampling_rate)
lbr.display.specshow(tonnetz, y_axis='tonnetz', x_axis='time')

plt.colorbar()
plt.title('Tonnetz Features')
plt.xlabel('Time')
plt.ylabel('Tonnetz Dimension')
plt.show()
```
